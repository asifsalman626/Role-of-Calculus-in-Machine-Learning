## **Loss Function Reduction Over Iterations**
**X-axis:** Number of Iterations
**Y-axis:** Loss Value
Shows how the model minimizes error over training.

Python Code to Visualize Loss Reduction

```python
import numpy as np
import matplotlib.pyplot as plt

iterations = np.arange(1, 101)
loss = np.exp(-iterations / 10) + 0.1 * np.random.rand(100)  # Simulated loss

plt.plot(iterations, loss, label="Loss over Iterations", color='blue')
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.title("Loss Reduction Over Time")
plt.legend()
plt.show()
```

# **Gradient Descent Path to Minimum**

**X-axis:** Parameter Value
**Y-axis:** Loss Function
Shows how gradient descent updates parameters to minimize the loss.

Python Code for Gradient Descent Visualizatio

```python
import numpy as np
import matplotlib.pyplot as plt

def loss_function(w):
    return w**2  # Simple quadratic function

w_values = np.linspace(-5, 5, 100)
loss_values = loss_function(w_values)

w_history = [4, 2, 1, 0.5, 0.1, 0]  # Simulated parameter updates
loss_history = [loss_function(w) for w in w_history]

plt.plot(w_values, loss_values, label="Loss Function", color='blue')
plt.scatter(w_history, loss_history, color='red', label="Gradient Descent Steps")
plt.xlabel("Parameter Value (w)")
plt.ylabel("Loss")
plt.title("Gradient Descent Optimization")
plt.legend()
plt.show()
```

# **Decision Boundary Evolution in Classification**

How a model improves its separation of classes over iterations.

 Python Code for Decision Boundary Evolution

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification

# Generate synthetic data with corrected parameters
X, y = make_classification(n_samples=100, n_features=2, n_informative=2,
                           n_redundant=0, n_repeated=0, n_classes=2, random_state=42)
model = LogisticRegression()

# Plotting decision boundaries over iterations
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, iterations in enumerate([1, 5, 20]):  # Different training stages
    model.set_params(max_iter=iterations)
    model.fit(X, y)
    ax = axes[i]

    # Plot data points
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='k')

    # Plot decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')

    ax.set_title(f"Decision Boundary After {iterations} Iterations")

plt.show()
```

# **Gradient Descent for Linear Regression**

Python code to implement gradient descent for linear regression.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a simple dataset (y = 2x + 1 + noise)
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 2 * X + 1 + np.random.randn(100, 1)

# Initialize parameters
w = np.random.randn(1)
b = np.random.randn(1)

# Hyperparameters
learning_rate = 0.1
iterations = 1000
m = len(X)  # Number of training samples

# Gradient Descent Function
def compute_gradients(X, y, w, b):
    y_pred = X * w + b
    error = y_pred - y
    # Compute the gradients for w and b
    dw = (2 / m) * np.sum(error * X)
    db = (2 / m) * np.sum(error)
    return dw, db

# Gradient Descent Loop
for i in range(iterations):
    dw, db = compute_gradients(X, y, w, b)

    # Update the parameters
    w -= learning_rate * dw
    b -= learning_rate * db

    # Optional: Print the loss every 100 iterations
    if i % 100 == 0:
        loss = np.mean((y - (X * w + b))**2)
        print(f"Iteration {i}: Loss = {loss}")

# Plotting the results
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, X * w + b, color='red', label='Fitted line')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Gradient Descent for Linear Regression')
plt.show()

# Final parameters
print(f"Final weight (w): {w}")
print(f"Final bias (b): {b}")
```

# **Python Implementation of a Simple CNN**

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Define a simple CNN model
model = keras.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

print(model.summary())  # Display model structure
```

**Python Implementation of a Simple RNN:**

```python
model = keras.Sequential([
    layers.SimpleRNN(50, activation='relu', input_shape=(10, 1)),
    layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')

print(model.summary())  # Display model structure
```
